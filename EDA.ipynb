```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#    JSON 
file_path = 'data/video-card.json'  #      
df = pd.read_json(file_path)

#     
print(df.head())

#    
print("\nDataFrame Info:")
df.info()

#    
print("\nStatistical Summary:")
print(df.describe())

#    
print("\nMissing Values:")
print(df.isnull().sum())

#  

# 1.   
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, color='blue', bins=10)
plt.title('Price Distribution of Video Cards')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()

# 2.   
plt.figure(figsize=(10, 6))
sns.histplot(df['memory'], kde=True, color='green', bins=10)
plt.title('Memory Size Distribution of Video Cards')
plt.xlabel('Memory (GB)')
plt.ylabel('Frequency')
plt.show()

# 3.    (, price, memory, core_clock, boost_clock)
corr = df[['price', 'memory', 'core_clock', 'boost_clock']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 4. Boxplot      
plt.figure(figsize=(10, 6))
sns.boxplot(x='color', y='price', data=df)
plt.title('Price Distribution by Color')
plt.xlabel('Color')
plt.ylabel('Price ($)')
plt.show()
```

```python
import pandas as pd
from rapidfuzz import fuzz, process

#  

#      
def get_best_match(name, choices, scorer=fuzz.ratio, cutoff=80):
    return process.extractOne(name, choices, scorer=scorer, score_cutoff=cutoff)

#       
matches = []
for name1 in df_cpu[' ']:
    match = get_best_match(name1, cpu_ratings[' '])
    if match:
        matches.append((name1, match[0]))

#   DataFrame
matches_df = pd.DataFrame(matches, columns=["Name Data1", "Best Match Data2"])

print(matches_df)
```

```python
merged_df = pd.merge(df_cpu, matches_df, left_on="name", right_on="Name Data1", how="left")

#     data2    "Best Match Data2"
final_merged_df = pd.merge(merged_df, cpu_ratings, left_on="Best Match Data2", right_on=" ", how="inner")
print(final_merged_df)
```

```python
final_merged_df.drop(columns=['Name Data1', 'Best Match Data2', ' _y', ' ', ' _x'], inplace=True)
```

```python
final_merged_df.to_json('data/cpu_merged.json', orient='records')
```
